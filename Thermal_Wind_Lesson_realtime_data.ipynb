### Background
The primitive equations:
1. Horizontal acceleration equation (F=ma):
$$ \frac{D}{Dt} \vec V_h = -f \hat k \times \vec V_h - \vec \nabla_p \Phi +\vec F_r  $$
or, in component form, 
$$ \frac{Du}{Dt} = fv  - \frac{\partial \Phi}{\partial x} +F $$
$$ \frac{Dv}{Dt} = -fu - \frac{\partial \Phi}{\partial y} +G $$

1. Vertical acceleration equation (aka Hydrostatic balance): 
$$ \frac {\partial \Phi}{\partial p} = - \alpha \equiv -\frac{1}{\rho} = \frac{RT}{p} $$
1. Mass continuity: 
$$ 0 = \frac {\partial u}{\partial x} + \frac {\partial v}{\partial y}  + \frac {\partial \omega}{\partial p} $$
1. First Law of thermodynamics: 
$$ \frac {D_h T}{Dt} + \omega \frac {\partial \theta}{\partial p} = J/C_p $$


Now suppose the flow is in geostrophic balance (dV/dt=0, F=G=0). This defines the geostrophic wind:  
$$ fv_g = +\frac{\partial \Phi}{\partial x}$$
$$ fu_g = -\frac{\partial \Phi}{\partial y}$$


Taking the p derivative, and noticing that $\frac{\partial \Phi}{\partial p} = g\frac{\partial Z}{\partial p}$ is the *geometric thickness* (dZ) of a layer with mass (dp/g), 
$$ \frac {\partial}{\partial p}v_g = +g/f \frac{\partial \frac {\partial Z}{\partial p}}{\partial x}$$
$$ \frac {\partial}{\partial p}u_g = -g/f \frac{\partial \frac {\partial Z}{\partial p}}{\partial y}$$

The horizontal pattern of thickness is the *pattern of temperature*, as you know. But actually plugging in the hydrostatic equation above for $\frac{\partial \Phi}{\partial p}$ is complicated, because p is in the denominator of the right hand side. ## Exercise 5 answer: 

In the troposphere the bottom of the jet is the warm side and the top of the jet is the cold side. Whereas in the stratosphere it is the opposite (top is warm side and bottom is cold side). This is the case due to the thermal wind equation. In the troposphere the reason temperature decreases with height is due to the fact that the geostrophic zonal wind component increases with height, using the thermal wind equation this makes dT/dy negative. In the stratosphere the zonal geostrophic wind component decreases with height, resulting in the thermal wind equation yielding a positive dT/dy.  

If the jet stream excursion into US latitude is viewed as a "vortex", it is considered a cold core in the tropopshere. Ther reasoning behind this is that there is cold air in the top side of the troposphere, meaning there is cold air in the higher altitudes. 

The relvort isosurface is the strongest between 800m and 1200m.
