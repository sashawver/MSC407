### Background and math

You will decompose the total wind into geostrophic and ageostrophic aspects: $\vec V = \vec V_g + \vec V_a$. 

* Horizontal acceleration equations (F=ma) for inviscid flow:
$$ \frac{du}{dt} = - \frac{\partial \Phi}{\partial x} + fv $$

$$ \frac{dv}{dt} = - \frac{\partial \Phi}{\partial y} - fu $$

* The geostrophic wind is defined as *the wind that would occur if the LHS were zero*: 

$$ 0 = - \frac{\partial \Phi}{\partial x} + fv_g $$

$$ 0 = - \frac{\partial \Phi}{\partial y} - fu_g $$

## Exercise 1 Answer:

$$ \frac{\partial}{\partial y}(0 = - \frac{\partial \Phi}{\partial x}(1/f) + v_g)$$
$$ 0 = -\frac{\partial^2 \Phi}{\partial y \partial x}1/f + \frac{\partial \Phi}{\partial x}\frac{\partial \ f}{\partial y} - \frac{\partial \ v_g}{\partial y} $$
$$\frac{\partial}{\partial y}(0 = - \frac{\partial \Phi}{\partial y}(1/f) - u_g)$$
$$ 0 = -\frac{\partial^2 \Phi}{\partial y \partial x}1/f - \frac{\partial \Phi}{\partial y}\frac{\partial \ f}{\partial x} - \frac{\partial \ u_g}{\partial x} $$
$$-\frac{\partial^2 \Phi}{\partial y \partial x}1/f - \frac{\partial \Phi}{\partial y}\frac{\partial \ f}{\partial x} - \frac{\partial \ u_g}{\partial x} = -\frac{\partial^2 \Phi}{\partial y \partial x}1/f + \frac{\partial \Phi}{\partial x}\frac{\partial \ f}{\partial y} - \frac{\partial \ v_g}{\partial y} $$
#### cancel out terms and assume f does not change with y and you end up with the horizontal divergence of the geostrophic wind 
$$\frac{\partial \ v_g}{\partial y} + \frac{\partial \ u_g}{\partial x} $$
#### Key points: the geostrophic wind is boring for 2 reasons: because it is not connected to vertical velocity $\omega$, and also because it would never change with time since its definition sets d/dt to zero. The implication is that ageostrophic wind, despite its relatively small magnitude, is especially *interesting*. However, the geostrophic wind is adequate to approximate the direction and speed of advection of various scalar fields (temperature, vorticity) in the atmosphere.
