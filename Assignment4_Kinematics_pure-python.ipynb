### Background
**Kinematics** is not dynamics, it is just a *description* of how 2D vector fields vary in 2D. 

* Assume LINEAR dependence of u,v on x,y 
    * This involves 4 coefficients $c_1, c_2, c_3, c_4$
* Also, add a constant (not variying with x,y) vector field U,V 

$$ u = U + c_1 x  +c_2 y$$
$$ v = V + c_3 x  +c_4 y$$

## Exercise 1 Answer:
* $ div = \frac{\partial u}{\partial x} + \frac{\partial v}{\partial y}= (0+c_1 + 0) + (0+ 0+ c_4) = c_1 + c_4$
* $ \zeta_{rel} = \frac{\partial v}{\partial x} - \frac{\partial u}{\partial y} = (0 + 0 + c_2) -(0+c_3+ 0) = c_2 - c_3 $
* $ def1 = \frac{\partial u}{\partial x} - \frac{\partial v}{\partial y} = (0 + c_1 + 0) - (0 + 0 +c_4) = c_1 - c_4$
* $ def2 = \frac{\partial v}{\partial x} + \frac{\partial u}{\partial y} = (0 + 0+ c_3) + (0 + 0 + c_2) = c_3 + c_2$

## Exercise 2: Create and display the basic fields
* First, execute a code cell setting up a coordinate system (x,y) from -10 to 10
* Next, execute the cell defining the basic "pure" div, vort, def1, def2, and U,V fields 
* Finally, display them. The first one has been done for you. You complete the work. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up a coordinate grid centered on 0,0
n = 10
x,y = np.mgrid[-n:n+1, -n:n+1]
```

```python
# Pure divergence of unit magnitude
udiv,vdiv = 1*x, 1*y

# Pure vorticity of unit magnitude
uvor,vvor = -1*y, 1*x

# Pure def1 of unit magnitude
udef1,vdef1 = 1*x, -1*y

# Pure def2 of unit magnitude
udef2,vdef2 = 1*y, 1*x

# Pure U and V
uU, vU = 1, 0
uV, vV = 0, 1
```

## Answer to Exercise 4: 

Vorticity plus various strengths in U can be described as moving a system in the y direction with increasing values of U. The value of U dictates where the center of rotation will be located.   

At lower levels (say 850mb), you would see more rotation (positive voricity) within the system. 

At upper levels (say 200mb), the westerly winds would begin to dominate the flow of the system and you would see less rotation.
