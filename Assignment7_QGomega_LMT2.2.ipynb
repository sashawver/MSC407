### Background

The sense of the QG omega equation, if second derivatives are treated like fancy negative signs and coefficients like static stability are treated as constants, is: 

** (Upward motion) = (Vorticity advection increasing w/ height) + (Warm advection) **

Typically, low-level vorticity advection is weak, so that one can simply interpret *upper level (like 500 or 200 hPa) vorticity advection* as an indicator of the first term. 


Notice that the QG omega equation is a ***diagnostic*** equation, not a **prognostic** equation with a time derivative in it. That is, the right-hand terms do not "lead to" upward motion, they are *equal to it, right now*. 

How did this equation come from our governing Primitive Equations set, which has a prognostic form: *d/dt( {winds, and thickness-related T} )  = {PGF&Coriolis, and physical heating rate Q} )*?

The physical concept behind the QG omega equation is, somehow, to state or assume or express the condition that *the flow remains balanced at all times* -- in other words, $\partial (imbalance)/ \partial t$ = 0. That is, the QG "(Upward motion)" in the equation above is *the hypothetical vertical motion field that exactly compensates the balance-destroying tendencies* of the terms on the right-hand side
